```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.utils import shuffle
```

```python
def compute_y(x, W, bias):
    # dreapta de decizie
    # [x, y] * [W[0], W[1]] + b = 0
    return (-x * W[0] - bias) / (W[1] + 1e-10)
```

```python
def plot_decision_boundary(X, y , W, b, current_x, current_y):
    x1 = -0.5
    y1 = compute_y(x1, W, b)
    x2 = 0.5
    y2 = compute_y(x2, W, b)
    # sterge continutul ferestrei
    plt.clf()
    # ploteaza multimea de antrenare
    color = 'r'
    if(current_y == -1):
        color = 'b'
    
    plt.ylim((-1, 2))
    plt.xlim((-1, 2))
    plt.plot(X[y == -1, 0], X[y == -1, 1], 'b+')
    plt.plot(X[y == 1, 0], X[y == 1, 1], 'r+')
    # ploteaza exemplul curent
    plt.plot(current_x[0], current_x[1], color+'s')
    # afisarea dreptei de decizie
    plt.plot([x1, x2] ,[y1, y2], 'black')
    plt.show(block=False)
    plt.pause(0.3)
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([-1, 1, 1, 1])
```

```python
W = np.random.random((2))
b = np.random.random((1))

print(W, b)
plot_decision_boundary(X, y, W, b, X[0], y[0])
```

```python
for e in range(30):
    X, y = shuffle(X, y)
    for t in range(4):
        y_hat = np.dot(X[t], W) + b # np.sum(X[t] * W)
        loss = (y_hat - y) * (y_hat - y) / 2
        W = W - 0.1 * (y_hat - y[t]) * X[t]
        b = b - 0.1 * (y_hat - y[t])
        if e % 10 == 0:
            plot_decision_boundary(X, y, W, b, X[t], y[t])
            plt.show()
```

```python
y_hat = np.sign(np.dot(X, W) + b)
print(y_hat)
print((y == y_hat).mean())
```
