```python
class BagOfWords:
    
    def __init__(self):
        self.vocab = {}
        self.words = []

    def build_vocabulary(self, data):
        idx = 0
        for sentence in data:
            for word in sentence:
                # word = word.lower()
                if word not in self.vocab:
                    self.vocab[word] = idx
                    idx += 1
                    # self.vocab[word] = len(self.words)
                    self.words.append(word)
        return len(self.words)

    def get_features(self, data):
        result = np.zeros((data.shape[0], len(self.words))) # N x VOCAB_SIZE
        for sentence_idx, sentence in enumerate(data):
            for word in sentence:
                # word = word.lower()
                if word in self.vocab:
                    result[sentence_idx, self.vocab[word]] += 1 # MAT[INDICE PROPIZITIE, INDICE CUVANT] += 1
        return result
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, Normalizer

def normalize_data(train_data, test_data, type=None):
    scaler = None
    if type == 'standard':
        scaler = StandardScaler()

    elif type == 'min_max':
        scaler = MinMaxScaler()

    elif type == 'l1' or type == 'l2':
        scaler = Normalizer(norm=type)

    if scaler is not None:
        scaler.fit(train_data)
        scaled_train_data = scaler.transform(train_data)
        scaled_test_data = scaler.transform(test_data)
        return scaled_train_data, scaled_test_data
    else:
        return train_data, test_data
```

```python
bow_model = BagOfWords()
bow_model.build_vocabulary(train_data) 

training_features, testing_features = bow_model.get_features(train_data), bow_model.get_features(test_data)

norm_train, norm_test = normalize_data(training_features, testing_features, type='l2')
```

```python
svm_training_data = np.load('data_svm/svm_train_data.npy', allow_pickle=True)
svm_training_labels = np.load('data_svm/svm_train_labels.npy')

for i in range(training_data.shape[0]):
    if training_labels[i] == 0:
        plt.plot(training_data[i, 0], training_data[i, 1], 'og')
    else:
        plt.plot(training_data[i, 0], training_data[i, 1], 'xr')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(svm_training_data, svm_training_labels, test_size=0.3)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
svc.fit(X_train, y_train)
# svc.predict(X_test)
svc.score(X_test, y_test)
```

```python
neg = np.argwhere(y_train < 1)
pos = np.argwhere(y_train >= 1)

plt.scatter(X_train[neg, 0], X_train[neg, 1])
plt.scatter(X_train[pos, 0], X_train[pos, 1])

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = svc.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])
# plot support svc
ax.scatter(svc.support_vectors_[:, 0], svc.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')

# C=0.1
```
